## METHODOLOGICAL INFORMATION
--------------------------
### Exploration of data
There are a number of columns with null values, the primary reason is suspected to be the house does not fulfill
criteria of that column, hence leaving it blank (e.g the house has no fireplace, hence fireplace quality was left blank)

A possible solution for the stated example will be to check if the houses has fireplaces or not, before deciding 
the score for the fireplace quality. (There is a option 'NA' in fireplace quality for houses with no fireplace)
Simliar issues will be resolved in the same way as well.
                                             
Other missing entries can be filled up with one of these methods:
    (i) Replace value with zero/'NA'
    (ii) Replace value with the mean/median score
    (iii) Use linear regression to predict the score with a closely related variable
    
    
After which, we start to use "LabelEncoder" from "sklearn preprocessing" to transform ordinal features to 
numberic representation and also "pd.get_dummies" to transform categorial features to binary features

Finally, using correlation and polynominal features, we will then select features for our models.
    

### Different approaches to the models used
Here we will discuss different approaches for each model.
All models uses LinearRegression, Lasso, Ridge, ElasticNet from sklearn.linear_model

Model 1:
Uses all features, without under going polynomial transformation, only standard scaler to reduce the quantum of the variables 
Followed by LinearRegression to get a benchmark score.
Then use Lasso to eliminated variables that are not useful.
Redo the model with the remaining variables and run a fresh modelling test with LinearRegression, Lasso, Ridge and ElasticNet
    to select the best model.

Model 2:
Same as above, but drop 'ID' and 'PID' in the model as they are selected in the final model but these are categorial
features with no real impact in reality

Model 3:
Based on correlation, this model only selects 10 features (before any transformation and dummy columns).

Model 4:
Based on correlation, this model chooses highly correlated variables from discrete and continous variables to do poly transformation
Then followed by correlated nominal categories and categorial categories to be fitted in the model.
The target score is also scaled down by np.log so that the model coefficient does not need to be huge after variables are standard scaled
Once the features are selected, the process of running different models are the same as Model 1.

Model 5:
Dropping more variables to create a simpler model from model 4

Model 6:
Simliar preprocessing from Model 4. Additionally, we also removed outliers and use LinearRegression to predict the
empty "Lot Frontage" scores instead of changing them to 0. Lastly, we also drop more unrelated variables.

Model 7:
Simliar to model 6, dropping all categorical variables to create an even simpler model
